## VISUALLIZING WORLD DATA 

### Total cases all over the world

```python
df_world_over_time = df_world.groupby('date')[['total_cases', 'new_cases', 'total_deaths', 'new_deaths']].sum().reset_index()
```

```python
df_world_over_time
```

```python
df_world_over_time['date'] = pd.to_datetime(df_world_over_time['date'])
```

```python
df_world_over_time.info()
```

```python
plt.figure(figsize=(20,10))
plt.plot(df_world_over_time['date'], df_world_over_time['total_cases'])
plt.xlabel('Years')
plt.ylabel('Confirmed Cases')
plt.title(' Total cases all over the world')
plt.xticks(rotation = 90)
```

### Total deathes cases evolution over time

```python
plt.figure(figsize=(20,10))
plt.plot(df_world_over_time['date'], df_world_over_time['total_deaths'])
plt.xlabel('Years')
plt.ylabel('Deaths Cases')
plt.title(' Deathes cases all over the world')
plt.xticks(rotation = 90)
```

```python
plt.figure(figsize=(20,10))
plt.plot(df_world_over_time['date'], df_world_over_time['new_cases'], c='r',lw=2)
plt.xlabel('Years')
plt.ylabel('New Cases')
plt.title(' New cases all over the world')
plt.xticks(rotation = 90)
```
